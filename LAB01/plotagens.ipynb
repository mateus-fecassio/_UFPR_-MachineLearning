```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_labels=['[10,10]','[10,20]','[20,10]','[20,20]','[80,30]','[90,50]','[100,100]','[110,60]','[120,70]','[140,70]']
```

```python
euclidiana = [0.894,0.904,0.905,0.922,0.917,0.924,0.917,0.922,0.924,0.923]
```

```python
manhattan = [0.894,0.904,0.905,0.922,0.917,0.924,0.917,0.922,0.924,0.923]
```

```python
minkowski = [0.894,0.904,0.905,0.922,0.917,0.924,0.917,0.922,0.924,0.923]
```

```python
chebyshev = [0.071,0.131,0.087,0.108,0.163,0.096,0.145,0.145,0.11,0.145]
```

```python
fig, ([ax1, ax2], [ax3, ax4]) = plt.subplots(2,2,figsize=(22, 15), sharex=True)

#fig.suptitle('Comparativo da acurcia entre o uso de diferentes mtricas de distncias para diferentes tamanhos de redimensionamento das imagens ')

ax1.plot(x_labels, euclidiana,'.-')
ax1.set_xticklabels(x_labels, rotation=45, ha='right')
ax1.set_ylabel('Acurcia Euclidiana')
ax1.grid()


ax2.plot(x_labels, manhattan, '.-')
ax2.set_xticklabels(x_labels, rotation=45, ha='right')
ax2.set_ylabel('Acurcia Manhattan')
ax2.grid()

ax3.plot(x_labels, minkowski, '.-')
ax3.set_xticklabels(x_labels, rotation=45, ha='right')
ax3.set_ylabel('Acurcia Minkowski')
ax3.set_xlabel('tamanho [x,y]')
ax3.grid()

ax4.plot(x_labels, chebyshev, '.-')
ax4.set_xticklabels(x_labels, rotation=45, ha='right')
ax4.set_ylabel('Acurcia Chebyshev')
ax4.set_xlabel('tamanho [x,y]')
ax4.grid()


fig.tight_layout()
plt.show()
fig.savefig('DISTANCIAS.png', dpi=fig.dpi)
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(x_labels, euclidiana)

ax.set(xlabel='tamanho [x,y]', ylabel='acurcia', title='Acurcia para diferentes tamanhos utilizando distncia Euclidiana')
ax.set_xticklabels(x_labels, rotation=45, ha='right')
ax.grid()

fig.savefig("euclidiana.png")
plt.show()
```
